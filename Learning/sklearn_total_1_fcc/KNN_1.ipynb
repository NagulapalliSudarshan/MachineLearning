```python
import numpy as np
import pandas as pd
from sklearn import neighbors,metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
titles = ['buying','maint','doors','persons','lug_boot','safety','class']
df = pd.read_csv("car.data",names=titles)
print(df)
```

```python
X = df[
    ['buying',
    'maint',
    'safety']
].values
y=df[['class']]
print(X,y)
print(type(y))
```

```python
# coversiion of data
# print(len(X[0]))
le = LabelEncoder()
for i in range(len(X[0])):
  print(X[:i])
  X[:,i] = le.fit_transform(X[:,i])
# print(X)
```

```python
# label mapping
label_mapping = {
    'unacc' : 0,
    'acc':1,
    'good': 2,
    'vgood':3
}
y['class'] = y['class'].map(label_mapping)
y = np.array(y)
print(y,type(y))
```

```python
knn_model = neighbors.KNeighborsClassifier(n_neighbors = 25,weights = 'uniform')
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
knn_model.fit(x_train,y_train)
y_predictions = knn_model.predict(x_test)
acc = metrics.accuracy_score(y_test,y_predictions)
print(acc)
```

```python
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data_iris = datasets.load_iris()
# print(data_iris.keys())
features = data_iris.data
labels = data_iris.target
```

```python
print(features.shape)
print(labels.shape)
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(features,labels,test_size=0.2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

#SVM

The support vector calculates the distance between the points and the line.
To optimize the line it creates a margin and the line can lie anywhere between the margins...the distance between the margins has to be the greatest

```python
# SVM
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(x_train,y_train)

print(svm_model)
```

```python
y_svm_predictions = svm_model.predict(x_test)
acc = accuracy_score(y_test,y_svm_predictions)
print(acc)
print("actual : ",y_test)
print("actual : ",y_svm_predictions)
```

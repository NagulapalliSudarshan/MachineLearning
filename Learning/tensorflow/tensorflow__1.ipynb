```python
%tensorflow_version 2.x
import tensorflow as tf
print(tf)
```

```python
# tensors
string = tf.Variable("this is a vector",tf.string)
number = tf.Variable(10,tf.int16)
floating = tf.Variable(10.48,tf.float64)
print(string,number,floating,sep="\n")
```

```python
# rank and shape
rank0_tensor = tf.Variable(1,tf.int8)
rank1_tensor = tf.Variable([1,2,3,4],tf.int32)
rank2_tensor = tf.Variable([[1,2],[3,4]],tf.int32)
print(rank0_tensor,rank1_tensor,rank2_tensor,sep="\n")
print("-"*50)
print(rank2_tensor.shape,rank1_tensor.shape,rank0_tensor.shape)                    # shape will tell the dimensions
print("-"*50)
print(tf.rank(rank2_tensor),tf.rank(rank1_tensor),tf.rank(rank0_tensor),sep="\n")  # rank will tell the highest branching
```

```python
# ones,zeros and reshape
tensor1 = tf.ones([3,3,1])
tensor2 = tf.reshape(tensor1,[3,1,3])
tensor3 = tf.reshape(tensor2,[3,-1])
tensor0 = tf.zeros([3,3,3])
print(tensor1)
print(tensor2)
print(tensor3)
print(tensor0)
```

```python
print(tensor0)
print(tf.reshape(tensor0,[9,-1]))
print(tf.reshape(tensor0,[3,-1]))
```
